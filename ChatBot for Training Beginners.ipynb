```python
import nltk
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def tokenize(sentence):
    return nltk.word_tokenize(sentence)
def stem(word):
    print()
    return stemmer.stem(word)
def bag_of_words(tokenized_sentence,all_words):
    tokenized_sentence = [stem(w) for w in tokenized_sentence]
    bag = np.zeros(len(all_words), dtype=np.float32)
    for index, w in enumerate(all_words):
        if w in tokenized_sentence:
            bag[index]=1.0
    return bag

#let's test this classification function
sentence = ['my', 'dog', 'is', 'nice', 'as', 'well', 'as', 'cat']
words= ['cat', 'flour']
bagg = bag_of_words(sentence, words)
print(bagg)  
```

```python
every_word =[]
tags=[]
patterns_with_tags=[]

for intent in intents:
    tag=intent.get('tag')
    tags.append(tag)
    for pattern in intent['patterns']:
        w=tokenize(pattern)
        every_word.extend(w)
        patterns_with_tags.append((w,tag))
        
words_to_ignore =['?','!',".",","]
every_word = [stem(w) for w in every_word if w not in words_to_ignore]
every_word= sorted(set(every_word))
tags = sorted(set(tags))
print(tags)
print(patterns_with_tags)
```

```python
X_train =[]
Y_train=[]

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(binary=True, ngram_range=(1,2))

for (pattern, tag) in patterns_with_tags:
    bag = bag_of_words(pattern, every_word)
    X_train.append(bag)
    label = float(tags.index(tag))
    
    Y_train.append(label)

print(X_train)
```

```python
import numpy as np
X_train = np.array(X_train)
Y_train = np.array(Y_train)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
```

```python
class ChatDataset(Dataset):
    def __init__(self):
        self.n_samples=len(X_train)
        self.x_data = X_train
        self.y_data = Y_train
    def __getitem__(self,index):
        return self.x_data[index], self.y_data[index]
    def __len__(self):
        return self.n_samples
```

```python
batch_size=8
workers = 0
hidden_size=8
output_size=len(tags)
input_size = len(X_train[0])
learning_rate = 0.001
num_epochs = 1000

dataset = ChatDataset()
train_loader = DataLoader(dataset = dataset, batch_size=batch_size, shuffle=True, num_workers=workers)
print(train_loader)
```

## Model

```python
class Neural(nn.Module):
    def __init__(self, input, hidden, num_classes):
        super(Neural, self).__init__()
        self.l1 = nn.Linear(input, hidden)
        self.l2 = nn.Linear(hidden, hidden)
        self.l3 = nn.Linear(hidden, num_classes)
        self.relu = nn.ReLU()
    def forward(self, x):
        out = self.l1(x)
        out = self.relu(out)
        out = self.l2(out)
        out = self.relu(out)
        out = self.l3(out)
        return out
```

```python
model = Neural(input_size, hidden_size, output_size) #tu moze byc problem
print(model)
```

```python
device = torch.device("cpu")
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    for(words, labels) in train_loader:
        words = words.to(device)
        labels = labels.to(device) #tu moze byc problem
         
        labels = labels.long()
        outputs = model(words) #tu moze byc problem

        loss = criterion(outputs,labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    if (epoch +1)%100 ==0:
        print('epoch', (epoch+1)/num_epochs, "loss = ", loss.item())
```

```python
our_data = {
    "model_state": model.state_dict(),
    "input_size": input_size,
    "output_size": output_size,
    "hidden_size": hidden_size,
    "all_words": every_word,
    "tags": tags
}

FILE = "data.pth"
torch.save(our_data, FILE)

print('training complete')
```
